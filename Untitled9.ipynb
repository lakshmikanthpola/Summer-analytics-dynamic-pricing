```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource

output_notebook()

# Load dataset
df = pd.read_csv("dataset.csv")

# Filter one parking lot (e.g., 'BHMBCCMKT01')
lot_id = 'BHMBCCMKT01'
lot_df = df[df['SystemCodeNumber'] == lot_id].copy()

# Fix timestamp parsing: day comes first (DD-MM-YYYY)
lot_df['Timestamp'] = pd.to_datetime(lot_df['LastUpdatedDate'] + ' ' + lot_df['LastUpdatedTime'], dayfirst=True)
lot_df = lot_df.sort_values(by='Timestamp')

# Reset index
lot_df.reset_index(drop=True, inplace=True)

# Extract features
occupancy = lot_df['Occupancy'].values
capacity = lot_df['Capacity'].iloc[0]

# Model parameters
base_price = 10
alpha = 5
prices = []

# Apply linear pricing logic
for i in range(len(occupancy)):
    occ_ratio = occupancy[i] / capacity
    if i == 0:
        price = base_price + alpha * occ_ratio
    else:
        price = prices[-1] + alpha * occ_ratio
    prices.append(price)

# Add to DataFrame
lot_df['Price'] = prices
lot_df['TimeStep'] = range(len(lot_df))

# Bokeh plot
source = ColumnDataSource(lot_df)

p = figure(title=f"Model 1: Linear Pricing for {lot_id}",
           x_axis_label='Time Step',
           y_axis_label='Value',
           width=800, height=400)

p.line(x='TimeStep', y='Price', source=source, line_width=2, color='navy', legend_label='Price')
p.scatter(x='TimeStep', y='Price', source=source, size=6, color='navy')

p.line(x='TimeStep', y='Occupancy', source=source, line_dash='dashed', line_color='green', legend_label='Occupancy')

p.legend.location = "top_left"
show(p)
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

# Load dataset
df = pd.read_csv("dataset.csv")

# Filter one parking lot for simplicity
lot_id = 'BHMBCCMKT01'
lot_df = df[df['SystemCodeNumber'] == lot_id].copy()

# Combine date and time into a single timestamp
lot_df['Timestamp'] = pd.to_datetime(
    lot_df['LastUpdatedDate'] + ' ' + lot_df['LastUpdatedTime'],
    dayfirst=True
)
lot_df = lot_df.sort_values(by='Timestamp').reset_index(drop=True)

# Replace missing values with defaults
lot_df['Occupancy'] = pd.to_numeric(lot_df['Occupancy'], errors='coerce').fillna(0)
lot_df['Capacity'] = pd.to_numeric(lot_df['Capacity'], errors='coerce').fillna(1)  # Avoid divide-by-zero
lot_df['QueueLength'] = pd.to_numeric(lot_df['QueueLength'], errors='coerce').fillna(0)
lot_df['TrafficConditionNearby'] = pd.to_numeric(lot_df['TrafficConditionNearby'], errors='coerce').fillna(1)
lot_df['IsSpecialDay'] = pd.to_numeric(lot_df['IsSpecialDay'], errors='coerce').fillna(0)

# Demand model weights
base_price = 10
alpha = 4
beta = 2
gamma = 1
delta = 3
lambda_ = 0.3
vehicle_type_weights = {"Car": 1.0, "Bike": 0.5, "Truck": 1.5}

# Map vehicle types to weights and fill unknowns
lot_df['VehicleWeight'] = lot_df['VehicleType'].map(vehicle_type_weights).fillna(1.0)

# Compute demand
occ_ratio = lot_df['Occupancy'] / lot_df['Capacity']
queue = lot_df['QueueLength']
traffic = lot_df['TrafficConditionNearby']
special_day = lot_df['IsSpecialDay']
vehicle_weight = lot_df['VehicleWeight']

demand = (
    alpha * occ_ratio +
    beta * queue -
    gamma * traffic +
    delta * special_day +
    vehicle_weight
)

# Normalize demand
demand_min = demand.min()
demand_max = demand.max()
normalized_demand = (demand - demand_min) / (demand_max - demand_min)

# Price = base * (1 +  * normalized_demand)
price = base_price * (1 + lambda_ * normalized_demand)
price = price.clip(lower=base_price * 0.5, upper=base_price * 2.0)

# Store outputs
lot_df['Demand'] = demand
lot_df['NormalizedDemand'] = normalized_demand
lot_df['Price_Model2'] = price
lot_df['TimeStep'] = range(len(lot_df))

# Plot
source = ColumnDataSource(lot_df)

p = figure(title=f"Model 2: Demand-Based Pricing for {lot_id}",
           x_axis_label='Time Step',
           y_axis_label='Price ($)',
           width=800, height=400)

p.line(x='TimeStep', y='Price_Model2', source=source, line_width=2, color='crimson', legend_label='Model 2 Price')
p.scatter(x='TimeStep', y='Price_Model2', source=source, size=6, color='crimson')

p.line(x='TimeStep', y='Occupancy', source=source, line_color='gray', line_dash='dotted', legend_label='Occupancy')

p.legend.location = "top_left"
show(p)
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import matplotlib.pyplot as plt

output_notebook()

# Load dataset and filter a single parking lot
df = pd.read_csv("dataset.csv")
lot_id = 'BHMBCCMKT01'
lot_df = df[df['SystemCodeNumber'] == lot_id].copy()

# Create unified timestamp
lot_df['Timestamp'] = pd.to_datetime(
    lot_df['LastUpdatedDate'] + ' ' + lot_df['LastUpdatedTime'],
    dayfirst=True
)
lot_df = lot_df.sort_values(by='Timestamp').reset_index(drop=True)

# Ensure correct datatypes
for col in ['Occupancy', 'Capacity', 'QueueLength', 'TrafficConditionNearby', 'IsSpecialDay']:
    lot_df[col] = pd.to_numeric(lot_df[col], errors='coerce')

# Clip occupancy to capacity (real-world logic)
lot_df['Occupancy'] = np.minimum(lot_df['Occupancy'], lot_df['Capacity'])

# Fill missing values with safe defaults
lot_df['QueueLength'] = lot_df['QueueLength'].fillna(0)
lot_df['TrafficConditionNearby'] = lot_df['TrafficConditionNearby'].fillna(1)
lot_df['IsSpecialDay'] = lot_df['IsSpecialDay'].fillna(0)

# Vehicle type mapping
vehicle_type_weights = {'Car': 1.0, 'Bike': 0.5, 'Truck': 1.5}
lot_df['VehicleWeight'] = lot_df['VehicleType'].map(vehicle_type_weights).fillna(1.0)

# Model parameters
base_price = 10
alpha = 4
beta = 1.5
gamma = 2
delta = 3
epsilon = 1.2
lambda_ = 0.5

# Demand calculation
occ_ratio = lot_df['Occupancy'] / lot_df['Capacity']
demand = (
    alpha * occ_ratio +
    beta * lot_df['QueueLength'] -
    gamma * lot_df['TrafficConditionNearby'] +
    delta * lot_df['IsSpecialDay'] +
    epsilon * lot_df['VehicleWeight']
)

# Improved normalization (center + stretch + scale to 01)
normalized_demand = (demand - demand.mean()) / demand.std()
normalized_demand = (normalized_demand - normalized_demand.min()) / (normalized_demand.max() - normalized_demand.min())

# Price calculation with bounding
price = base_price * (1 + lambda_ * normalized_demand)
price = price.clip(lower=base_price * 0.5, upper=base_price * 2.0)

# Add to DataFrame
lot_df['Demand'] = demand
lot_df['NormalizedDemand'] = normalized_demand
lot_df['Price_Model2'] = price
lot_df['TimeStep'] = range(len(lot_df))

# Diagnostic printout
print("Demand Diagnostic Report")
print("-------------------------")
print("Raw Demand Range:       ", round(demand.min(), 2), "to", round(demand.max(), 2))
print("Normalized Demand Range:", round(normalized_demand.min(), 2), "to", round(normalized_demand.max(), 2))
print("Final Price Range:      ", round(price.min(), 2), "to", round(price.max(), 2))

# Debug plot using Matplotlib
plt.figure(figsize=(10, 4))
plt.plot(demand, label="Raw Demand", color='black')
plt.plot(normalized_demand, label="Normalized Demand", color='blue')
plt.plot(price, label="Model 2 Price", color='red')
plt.title("Model 2: Demand to Price Debug View")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()

# Bokeh plot
source = ColumnDataSource(lot_df)

p = figure(title=f"Model 2: Demand-Based Pricing for {lot_id}",
           x_axis_label='Time Step',
           y_axis_label='Price ($)',
           width=800, height=400)

p.line(x='TimeStep', y='Price_Model2', source=source, line_width=2, color='crimson', legend_label='Model 2 Price')
p.scatter(x='TimeStep', y='Price_Model2', source=source, size=6, color='crimson')

p.line(x='TimeStep', y='Occupancy', source=source, line_color='gray', line_dash='dotted', legend_label='Occupancy')

p.legend.location = "top_left"
show(p)
```

```python
# Ensure all inputs are numeric
df['Occupancy'] = pd.to_numeric(df['Occupancy'], errors='coerce').fillna(0)
df['Capacity'] = pd.to_numeric(df['Capacity'], errors='coerce').fillna(1)
df['QueueLength'] = pd.to_numeric(df['QueueLength'], errors='coerce').fillna(0)
df['TrafficConditionNearby'] = pd.to_numeric(df['TrafficConditionNearby'], errors='coerce').fillna(1)
df['IsSpecialDay'] = pd.to_numeric(df['IsSpecialDay'], errors='coerce').fillna(0)
df['VehicleWeight'] = df['VehicleType'].map({"Car": 1.0, "Bike": 0.5, "Truck": 1.5}).fillna(1.0)

# Compute Model 2 pricing
df['OccRatio'] = df['Occupancy'] / df['Capacity']
df['Demand'] = (
    4 * df['OccRatio'] +
    1.5 * df['QueueLength'] -
    2 * df['TrafficConditionNearby'] +
    3 * df['IsSpecialDay'] +
    1.2 * df['VehicleWeight']
)
df['NormalizedDemand'] = (df['Demand'] - df['Demand'].mean()) / df['Demand'].std()
df['NormalizedDemand'] = (df['NormalizedDemand'] - df['NormalizedDemand'].min()) / (df['NormalizedDemand'].max() - df['NormalizedDemand'].min())

df['Price_Model2'] = 10 * (1 + 0.5 * df['NormalizedDemand'])
df['Price_Model2'] = df['Price_Model2'].clip(lower=5, upper=20)
```

```python
from math import radians, sin, cos, sqrt, atan2

# Distance function (Haversine)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Find unique lot positions
lots = df[['SystemCodeNumber', 'Latitude', 'Longitude']].drop_duplicates().reset_index(drop=True)

# Build neighbors dictionary: for each lot, store nearby lot IDs within 1 km
neighbors = {}
for i, row in lots.iterrows():
    lot_id = row['SystemCodeNumber']
    lat1, lon1 = row['Latitude'], row['Longitude']
    nearby = []
    for j, other in lots.iterrows():
        if lot_id == other['SystemCodeNumber']:
            continue
        lat2, lon2 = other['Latitude'], other['Longitude']
        dist = haversine(lat1, lon1, lat2, lon2)
        if dist <= 1.0:
            nearby.append(other['SystemCodeNumber'])
    neighbors[lot_id] = nearby

# : adjustment strength
eta = 0.3
base_price = 10

# Initialize Model 3 with Model 2 prices
df['Price_Model3'] = df['Price_Model2']

# Apply competitive adjustment for each row
for idx, row in df.iterrows():
    lot_id = row['SystemCodeNumber']
    time = row['Timestamp']
    own_price = row['Price_Model2']

    nearby_ids = neighbors.get(lot_id, [])
    if not nearby_ids:
        continue

    # Get nearby lots at the same timestamp
    competitors = df[
        (df['SystemCodeNumber'].isin(nearby_ids)) &
        (df['Timestamp'] == time)
    ]

    if competitors.empty:
        continue

    avg_nearby_price = competitors['Price_Model2'].mean()
    adjusted_price = own_price + eta * (avg_nearby_price - own_price)
    adjusted_price = max(base_price * 0.5, min(base_price * 2.0, adjusted_price))

    df.at[idx, 'Price_Model3'] = adjusted_price

print(" Model 3 pricing applied successfully.")
print("Sample Price Range:", df['Price_Model3'].min(), "to", df['Price_Model3'].max())
```
